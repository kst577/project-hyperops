```python
import numpy as np
import random
import pandas as pd
```

```python
hours=[]
for i in range(1,19):
    a=int(input('enter the passenger flow at hour {} '.format(i)))
    hours.append(a)


#taking 18 hours operations as inputs
```

```python
def rd(n, total_sum):
    nums = np.random.rand(n)
    return nums/np.sum(nums)*total_sum


# function inorder to randomly distrubute the hourly flow to 13 platforms
```

```python
# Defining empty lists

group1_full_operations=[]
group2_full_operations=[]
group3_full_operations=[]
group4_full_operations=[]
group5_full_operations=[]
group6_full_operations=[]



for i in hours:
   
    a=rd(13,i) # function calling
    b=a.tolist()
    c=[int(i) for i in b]
    # Assigning the number of people to respected platforms
    
    p1=c[0]
    p2=c[1]
    p3=c[2]
    p4=c[3]
    p5=c[4]
    p6=c[5]
    p7=c[6]
    p8=c[7]
    p9=c[8]
    p10=c[9]
    p11=c[10]
    p12=c[11]
    p13=c[12]
     
    group1_full_operations.append(p1+p2+p3)
    group2_full_operations.append(p4+p5)
    group3_full_operations.append(p6+p7)
    group4_full_operations.append(p8+p9)
    group5_full_operations.append(p10+p11)
    group6_full_operations.append(p12+p13)
    
              
```

```python
time=['6:00 AM','7:00 AM','8:00 AM','9:00 AM','10:00 AM','11:00 AM','12:00 PM',
     '1:00 PM','2:00 PM','3:00 PM','4:00 PM','5:00 PM','6:00 PM','7:00 PM','8:00 PM',
     '9:00 PM','10:00 PM','11:00 PM']
```

```python
df=pd.DataFrame({'Hour':time,'Group 1':group1_full_operations,'Group 2':group2_full_operations,
                'Group 3':group3_full_operations,'Group 4':group4_full_operations,
                'Group 5':group5_full_operations,'Group 6':group6_full_operations})
```
